## Step 0: Converting .csv File to .db File

To make this project as realistic as possible, we need to get our data from a proper SQL server. However, I don't have any, so as a solution we can convert and use a .csv file as .db file. This way we can use SQlAlchemy library to get the data form .db file, and create a proper SQl qeries to connect, get data and disconnect from a SQL server. 

```python
#This is an hourly electricity consumption data with 8784 records for the whole 2016 year. 
import sqlalchemy.types
import pandas as pd
import sqlalchemy

import numpy as np

data = pd.read_csv('energy_consumption_levels.csv')

# Try to declare column names and data types by hand. (Names does appear on the .db table but not the dtypes.)
columns = ("3_levels","5_levels","7_levels","consumption","temperature","hour_of_day","day_of_week","day_of_month","month_of_year")           
dtypes = (sqlalchemy.types.SMALLINT, sqlalchemy.types.SMALLINT, sqlalchemy.types.SMALLINT, sqlalchemy.types.FLOAT(8), sqlalchemy.types.FLOAT(8), sqlalchemy.types.SMALLINT, sqlalchemy.types.SMALLINT, sqlalchemy.types.SMALLINT, sqlalchemy.types.SMALLINT)
# The to_sql() function requies them as dict.
names = dict(zip(columns,dtypes))

# Creating a new, empty .db file 
engine = sqlalchemy.create_engine('sqlite:///energy_consumption_levels.db')

data.to_sql('energy', engine, if_exists='replace', index=False, dtype = names)

# Closing the connection. (Normally you would use "connection = engine.connect()" and "connection.close()" it wil be shown later. )
engine.dispose()
```

## Step1: Loading Data from SQL Server with SQLAlchemy

Alright, now we can start. First we get the data from a locally existing file. This is a bit less realistic compared to getting the data from a proper SQL server. Because we should get the connection URL from a .env file for a real SQl connection. Anyway, let's focus on to get the dataset form .db file.  

```python
engine = sqlalchemy.create_engine('sqlite:///energy_consumption_levels.db')

connection = engine.connect()

# Pushing the SQL query to get the data from the .db file, and converting them to pandas.Dataframe(). 
energy_data = pd.DataFrame(engine.execute('SELECT * FROM energy').fetchall(), columns = engine.execute('SELECT * FROM energy').keys())

connection.close()

energy_data.head()
```

This is the data we have. It has a little more than what we want so, the next step is to clean and organize to give it a proper shape.

## Step 2: Cleaning the Data

Now we can clean and select the useful information inside our data. To begin, we should trim away unnecessary columns. For instance, '3_levels' and '5_levels' won't be used later on so we need to get rid of them. 

### 2.1 Organising The Dataset

The problem with this part is if you executed the lines below and you tried to execute them again, the code will not work. It will give an error with the main concept of "the columns you specified does not exist". 

```python
energy_data.drop(columns = ["3_levels","5_levels","7_levels"], inplace= True)
```

Now the "levels" are out, we need to organise the time values. They were 4 different columns for hour, day, month and specific day of the week. Let's start with the "day_of_week".

```python
# Since the 'day_of_week' column contains numbers 1 to 7, we should map these values to Monday through Sunday.
day_mapping = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}
# Changing the dtype to int in order to ensure that every datum is mapable to Monday through Sunday.
energy_data['day_of_week'].astype(int)

energy_data['day_of_week'] = energy_data['day_of_week'].replace(day_mapping)
```

```python
# Updating the column names with more understandable headings. 
energy_data.rename(columns={'month_of_year': 'month', 'day_of_month' : 'day', 'hour_of_day' : 'hour'}, inplace=True)
energy_data.head()
```

Now this looks a lot more pretty than what we started with but, it still has its problems. Let's try to combine the time values properly.

```python
# In order to change the 1-24 hour scale with "pd.to_datetime()" function, we need to make it 0-23 first.
for i in range(0,len(energy_data['hour'])):
    if energy_data['hour'][i] == 24:
        energy_data['hour'][i] = 0

# Also without the year value, "pd.to_datetime()" does not work the way we want. 
energy_data['year'] = 2016    
combined_time = energy_data[['year', 'month', 'day', 'hour']]
combined_time.head()
```

Now we convert it to the "datetime" and clean the unnecessary columns. 

```python
energy_data['combined_datetime'] = pd.to_datetime(combined_time, format='%Y-%m-%d %H')
```

```python
energy_data.drop(columns = ["hour","day","month","year"], inplace= True)
energy_data.head()
```

### 2.2 Cleaning The Dataset

Now we have our data organized but, does this have any NaN or Nat values? Does this have any data that can mislead our AI models? Let's take a look to the data graph and see. 

```python
# Please run this part twice to see the graph.
import matplotlib.pyplot as plt

energy_data.dropna() #It does not have any NaN or NaT values luckily. (If there are few and you have millions of data, maybe consider just dropping them.)
# Keep that in mind, sometimes NaN and NaT values gives important information about the situation.

plt.figure(figsize=(18,6))
plt.plot(energy_data["combined_datetime"],energy_data["consumption"])
plt.show()
```

Overall, it looks clean and workable but if you zoom a little closer you can see anomalies. Before we try to detect them, first we need to make the Exploratory Analysis. This means we need to look at the Seasonality, Trends and Autocoorilations. 

#### Exploratory Analysis

Let's look at the stationality first. In data analysis, it is important to use stationary data to make correct inferences. If a dataset is stationary then we can use our autocoorilation functions and AI/ML models more accurately than unfiltered data. It is not hard to detect and eleminate seasonality and treds but it is a little bit challenging concept in mathematics and statistics. You can take a look at this subject here. 

We will analyse the dataset using the Augmented Dickey Fuller Test. You can take a look at this subject here [<sup>[1]</sup>] .

[1]: https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test

```python
from statsmodels.tsa.stattools import adfuller 

adf_result = adfuller(energy_data["consumption"])

if adf_result[0] < adf_result[4]['1%']:
    print(f"The dataset is stationary, because the p values is smaller than critical value. P-value: {adf_result[1]:.2g}")
elif adf_result[0] > adf_result[4]['1%']:
    print(f"The dataset is non-stationary, because the p values is bigger than critical value. P-value: {adf_result[1]:.2g}")
    
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```

Now we know our dataset is stationary, we can take a look at the seasonality and trend. The mathematical background is also here [<sup>[2]</sup>] .

[2]: https://machinelearningmastery.com/decompose-time-series-data-trend-seasonality/

```python
from statsmodels.tsa.seasonal import seasonal_decompose

sd_result = seasonal_decompose(energy_data["consumption"], freq=2000)

plt.figure(figsize=(18, 16))
plt.subplot(411)
plt.plot(energy_data["consumption"], label='Consumption')

plt.subplot(412)
plt.plot(sd_result.trend.index, sd_result.trend, label='Trend')
plt.legend()

# Plotting the seasonal component
plt.subplot(413)
plt.plot(sd_result.seasonal.index, sd_result.seasonal, label='Seasonal')
plt.legend()

# Plotting the residual component
plt.subplot(414)
plt.plot(sd_result.resid.index, sd_result.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

As you can see, the data has trend. But there is a problem. The conscumption rates does not go higher in the winter like you would expect, even summer months has more energy consumptions. In order to understand the data fully we need to look at the source of the data. 

The data we have is the energy consumption of a commercial-type house located in Taiwan. Taiwan is in the subtropical climate zone, which indicates that Taiwan has hot summers and mild winters as a characteristic trait [<sup>[3]</sup>] [<sup>[4]</sup>] . This situation gives more clearity to the energy consumption graph.



But in temparature graph we see -20 degree Celcius, this surely means cold. This might not be reflect felt cold inside the house, because the tempereture data comes from meteorological sensors of a specialized institute.

#### Autocorrelation Tests

In these tests, we measure the relationship of the individual data in our data set with the data that came before it. Especially in the autocorrelation test chart, we can see the connection of each data with the previous data. The difference between autocorrelation and partial autocorrelation comes from their mathematical foundations. To explain briefly, partial autocorrelation looks at the relationship between purer and independent data by removing old data pieces from our original data that indirectly affect the values of the data. Since it is a difficult subject to explain and understand, you can get help from additional resources here [<sup>[5]</sup>] [<sup>[6]</sup>] .

[5]: https://www.ibm.com/docs/en/spss-modeler/18.2.0?topic=data-autocorrelation-partial-autocorrelation-functions
[6]: https://www.youtube.com/watch?app=desktop&v=DeORzP0go5I


```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(energy_data["consumption"], lags=30) 
plt.title('Autocorrelation Function (ACF)')
plt.show()

plot_pacf(energy_data["consumption"], lags=30) 
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

Since the data is stored hourly, it is highly related to the data 24 hours before. However, if we decide to remove the indirect effect, we can see that the current data is mostly related with 2 previous data and 25th datum (0th datum is the datum itself, so it is always 1).  

So what does this tells us? It means that the data set is consistent within itself and has a trend. We can build a forcasting algorithm to guess the next value with the data we have it will have a solid background. In different datasets, these tests can give you way more information about the situation's behavior and help you understand what the data says to you. 

#### Detection Algorithms

We will use some statistics-based algorithms, and look at their intersection points for anomalies. We will use Rolling Z-score,  MAD Rolling z-score, IQR, Laplace Distribution, Isolation Forest, and DBSCAN. We can use different algorithms for anomaly detection but I choose them like this. 

Specifically, I did not use any ML-based algorithms. Because having an ML architecture for each data-generating machine might be problematic in performance vise. 

```python
# Rolling Z-Score
window_size = 24 # Rolling z-score is just the z-score algorithim but with the moving window. 
con_mean = energy_data["consumption"].rolling(window=window_size).mean()
con_std = energy_data["consumption"].rolling(window=window_size).std()

z_score = (energy_data["consumption"] - con_mean)/con_std

z_score_anomalies = energy_data["consumption"][(z_score > 1.2) | (z_score < -1.6)] # You can finetune the thresholds via graph

plt.figure(figsize=(18,6))
plt.plot(energy_data["consumption"])
plt.plot(z_score_anomalies,".")
plt.show()
```

```python
# MAD Rolling Z-Score
# This has the same logic but different implementation. 

con_median = energy_data["consumption"].rolling(window=window_size).median()
con_mad = (abs(energy_data["consumption"] - con_median)).median()

rolling_mad_z_score = (energy_data["consumption"] - con_median) / (1.4826 * con_mad)
rolling_mad_z_score_anomalies = energy_data["consumption"][(rolling_mad_z_score > 1.1) | (rolling_mad_z_score < -2.7)]


plt.figure(figsize=(18,6))
plt.plot(energy_data["consumption"])
plt.plot(rolling_mad_z_score_anomalies,".")
plt.show()
```
